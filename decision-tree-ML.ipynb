# DECISION TREE

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import  accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/vendas.csv')
```

```python
# Convertendo os dados da coluna `Horario Clique` para um tipo numérico
# Str -> date -> numeric
df['Horario Clique'] = pd.to_numeric(pd.to_datetime(df['Horario Clique']))
```

```python
# Convertendo dados strings das colunas `Pais`, `Cabecalho` e `Cidade` para um tipo numérico
encoder = OrdinalEncoder()
df[['Pais', 'Cabecalho', 'Cidade']] = encoder.fit_transform(df[['Pais', 'Cabecalho', 'Cidade' ]])
```

```python
data = df.drop('Comprou', axis=1).values
data_class = df['Comprou']

x_treino, x_teste, y_treino, y_teste = train_test_split(data, data_class, test_size=0.3, random_state=3)
```

```python
tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(x_treino, y_treino)

# lista com os dados do teste
result = tree.predict(x_teste)

result
```

```python
accuraty = accuracy_score(y_teste, result)
accuraty
```

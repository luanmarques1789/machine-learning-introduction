# NAIVE BAYES

## Imports

```python
import pandas as pd

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Manipulating data

```python
from google.colab import drive
drive.mount('/content/drive')

local = '/content/drive/MyDrive/Colab Notebooks/Machine Learning/vendas.csv'
origin_df = pd.read_csv(local)

df = origin_df.drop(['Cabecalho', 'Cidade', 'Pais', 'Horario Clique'], axis=1)
```

```python
dados = df.drop('Comprou', axis=1)
target = df['Comprou']

dados_treino, dados_teste, target_treino, target_teste = train_test_split(dados, target, test_size=0.3, random_state=3)
```

## Score

```python
nb = GaussianNB()
nb.fit(dados_treino, target_treino)

accuracy = nb.score(dados_teste, target_teste)
accuracy
```

```python
# Classification report
print(classification_report(target_teste, nb.predict(dados_teste)))
```
